```python
df_cleaned = df2.dropna(subset=['booking_id'])
df_cleaned
```

```python
#1. Find the number of distinct bookings,sessions,and searches from the given datasets.

num_distinct_bookings = df['booking_id'].nunique()
num_distinct_sessions = df_cleaned['session_id'].nunique()
num_distinct_searches = df_cleaned['search_id'].nunique()

# Print the results
print(f'Number of distinct bookings: {num_distinct_bookings}')
print(f'Number of distinct sessions: {num_distinct_sessions}')
print(f'Number of distinct searches: {num_distinct_searches}')
```

```python
# 2. How many sessions have more than one booking?
session_bookings_count = df_cleaned.groupby('session_id')['booking_id'].count()

sessions_with_multiple_bookings = session_bookings_count[session_bookings_count > 1]

num_sessions_with_multiple_bookings = sessions_with_multiple_bookings.count()
# Print the result
print(f'Number of sessions with more than one booking: {num_sessions_with_multiple_bookings}')
```

```python
# 3. Which days of the week have the highest number of bookings? Also,draw a pie chart to show the distribution for all days of the week?

df['booking_time'] = pd.to_datetime(df['booking_time'])
df['day_of_week'] = df['booking_time'].dt.day_name()

bookings_by_day = df['day_of_week'].value_counts()
print("Bookings by day of the week:\n", bookings_by_day)
```

```python

# Plotting a pie chart for bookings distribution by day of the week

plt.figure(figsize=(8, 6))
plt.pie(bookings_by_day, labels=bookings_by_day.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of bookings by day of the week')
plt.show()
```

```python
# 4. For each of the service names, display the total number of bookings and the total Gross Booking Value in INR

service_summary = df.groupby('service_name').agg({'booking_id': 'count', 'INR_Amount': 'sum'})
service_summary.columns = ['Total Bookings', 'Total GBV (INR)']
print("Summary for each service name:\n", service_summary)
```

```python
# 5.For customers who have more than 1 booking,which is the most booked route (from_cityto to_city)?

multiple_bookings = df['customer_id'].value_counts()[df['customer_id'].value_counts() > 1].index

filtered_data = df[df['customer_id'].isin(multiple_bookings)]
print(filtered_data)
```

```python

most_booked_route = filtered_data.groupby(['from_city', 'to_city']).size().idxmax()

print("Most booked route (from_city to to_city):", most_booked_route)
```

```python
# 6. Which are the top 3 departure cities from where customers book mostly in advance, provided that there have been atleast 5 departures from that city?

in_advance_threshold = 3

bookings_in_advance = df[df['days_to_departure'] >= in_advance_threshold]

advance_bookings_count = bookings_in_advance['from_city'].value_counts()

advance_bookings_count_filtered = advance_bookings_count[advance_bookings_count >= 5]

top_departure_cities = advance_bookings_count_filtered.sort_values(ascending=False)

# Select the top 3 departure cities
top3_departure_cities = top_departure_cities.head(3)

print("Top 3 departure cities with at least 5 departures and bookings made in advance:")
print(top3_departure_cities)

```

```python
# 7.) Plot a heatmap displaying correlations of the numerical column and report whichpairofnumericalcolumnsinthebookingsdataset,havethemaximumcorrelation?
import seaborn as sns
```

```python

numerical_df = df[['INR_Amount', 'no_of_passengers', 'days_to_departure', 'distance_km']].dropna()


correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()


max_corr = correlation_matrix.unstack().sort_values(ascending=False)

max_corr = max_corr[max_corr < 1]
max_corr_pair = max_corr.idxmax()

print("Pair of numerical columns with the maximum correlation:")
print(f"{max_corr_pair[0]} and {max_corr_pair[1]} with a correlation of {max_corr[max_corr_pair]:.2f}")
```

```python
# 8)For each service,which is the most used device type for making bookings on the platform?
device_usage = df.groupby(['service_name', 'device_type_used']).size()

most_used_device_per_service = device_usage.groupby(level=0).idxmax().apply(lambda x: x[1])

print("Most used device type for making bookings on the platform for each service:")
print(most_used_device_per_service)
```

```python
# 9)Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a time series for each year and quarter showing the number of bookings performed by each device type.

df['booking_time'] = pd.to_datetime(df['booking_time'])


df['year'] = df['booking_time'].dt.year
df['quarter'] = df['booking_time'].dt.quarter


booking_trends = df.groupby(['year', 'quarter', 'device_type_used']).size().reset_index(name='num_bookings')

plt.figure(figsize=(12, 6))
for device_type, data in booking_trends.groupby('device_type_used'):
    plt.plot(data['year'].astype(str) + 'Q' + data['quarter'].astype(str), data['num_bookings'], label=device_type)
    
plt.title('Trends of Number of Bookings by Device Type (Quarterly)')
plt.xlabel('Year and Quarter')
plt.ylabel('Number of Bookings')
plt.legend(title='Device Type')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 10.
df['booking_time'] = pd.to_datetime(df['booking_time'])

# Extract month and day of the week
df['month'] = df['booking_time'].dt.month
df['day_of_week'] = df['booking_time'].dt.dayofweek

# Group by month and day of the week, count the number of searches and bookings for each group
grouped = df.groupby(['month', 'day_of_week'])
total_searches = grouped.size()
total_bookings = grouped['booking_id'].count()

# Calculate BSR for each group
bsr = total_bookings / total_searches

# Average BSR for each month of the year
average_bsr_month = bsr.groupby('month').mean()

# Average BSR for each day of the week
average_bsr_day_of_week = bsr.groupby('day_of_week').mean()

print("Average BSR for each month of the year:")
print(average_bsr_month)
print("\nAverage BSR for each day of the week:")
print(average_bsr_day_of_week)
```
